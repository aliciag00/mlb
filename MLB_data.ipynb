```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Group data by player and calculate total hits
player_hits = df.groupby('Player Name')['H'].sum()

# Sort players by hits
player_hits_sorted = player_hits.sort_values(ascending=False)

# Get top players with the most hits (e.g., top 10)
top_players = player_hits_sorted.head(10)

# Plot bar chart
plt.figure(figsize=(10, 6))
top_players.plot(kind='bar', color='skyblue')
plt.title('Players with the Most Career Hits')
plt.xlabel('Player Name')
plt.ylabel('Total Hits')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Group data by player and calculate total hits
team_hits = df.groupby('Team')['H'].sum()

# Sort players by hits
team_hits_sorted = team_hits.sort_values(ascending=False)

# Get top players with the most hits (e.g., top 10)
top_teams = team_hits_sorted.head(10)

# Plot bar chart
plt.figure(figsize=(10, 6))
top_teams.plot(kind='bar', color='skyblue')
plt.title('Teams with the Most Hits')
plt.xlabel('Team')
plt.ylabel('Total Hits')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Filter the DataFrame for a specific team (e.g., team 'NYA')
team_df = df[df['Team'] == 'Los Angeles Dodgers']
season_df = df[df['Season'] == '2018']

# Plot scatter plot
sns.scatterplot(x="weight", y="HR", data=team_df)
plt.title('Weight vs HR for Los Angeles Dodgers')
plt.xlabel('Weight')
plt.ylabel('HR')
plt.show()
```

```python
# Add a new column with the result of dividing one column by another
df['AVG'] = df['H'] / df['AB']
```

```python
season_df = df[df['Season'] == '2018']
team_df = df[df['Team'] == 'Los Angeles Dodgers']

# Plot scatter plot
sns.scatterplot(x="AVG", y="salary", data=team_df)
plt.title('AVG vs Salary for 2018 Los Angeles Dodgers')
plt.xlabel('AVG')
plt.ylabel('salary')
plt.show()
```

```python
# Group data by player and calculate total hits
team_salary = df.groupby('Team')['salary'].sum()

# Sort players by hits
team_salary_sorted = team_salary.sort_values(ascending=False)

# Get top players with the most hits (e.g., top 10)
top_team_salary = team_salary_sorted.head(10)

# Plot bar chart
plt.figure(figsize=(10, 6))
top_team_salary.plot(kind='bar', color='skyblue')
plt.title('Teams with the Highest Salaries')
plt.xlabel('Team')
plt.ylabel('Total Salaries')
plt.xticks(rotation=45, ha='right')
plt.show()
```
